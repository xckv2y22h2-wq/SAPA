# Caltech 101 

```python
import os
import torch
from torchvision import datasets, transforms
import matplotlib.pyplot as plt

# Set dataset path and choose which dataset to load
data_path = '/media/tom/datasets'
dataset_name = 'caltech101'  # or 'cifar10'

# Define image transform
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

# Load dataset
if dataset_name == 'caltech101':
    dataset = datasets.Caltech101(root=data_path, download=False, transform=transform)
elif dataset_name == 'cifar10':
    dataset = datasets.CIFAR10(root=data_path, train=True, download=True, transform=transform)
else:
    raise ValueError("Unsupported dataset.")

# Show dataset details
print(f"Dataset: {dataset_name}, total samples: {len(dataset)}")

# Show one image and label
img, label = dataset[0]
plt.imshow(img.permute(1, 2, 0))  # Convert (C,H,W) to (H,W,C)
plt.title(f"Label: {dataset.categories[label]}")
plt.axis("off")
plt.show()
```

# FGVC Aircraft 

```python
import os
import torch
from torchvision import datasets, transforms
import matplotlib.pyplot as plt

# Set your desired path
data_path = '/media/tom/datasets'

# Define transform
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

# Load FGVC Aircraft dataset (train split)
dataset = datasets.FGVCAircraft(root=data_path, split='train', download=True, transform=transform)

# Check dataset size
print(f"FGVC Aircraft - Total samples: {len(dataset)}")
print(f"Number of classes: {len(dataset.classes)}")

# Show one image and label
img, label = dataset[0]
plt.imshow(img.permute(1, 2, 0))  # (C, H, W) → (H, W, C)
plt.title(f"Label: {dataset.classes[label]}")
plt.axis("off")
plt.show()
```

# SUN 397 

```python
import os
import torch
from torchvision import datasets, transforms
import matplotlib.pyplot as plt

# Dataset setup
data_path = '/media/tom/datasets'
dataset_name = 'sun397'  # or 'caltech101', 'cifar10'

# Define transform
transform = transforms.Compose([
    transforms.Resize((224, 224)),
    transforms.ToTensor()
])

# Load dataset
if dataset_name == 'sun397':
    dataset = datasets.SUN397(root=data_path, download=True, transform=transform)
elif dataset_name == 'caltech101':
    dataset = datasets.Caltech101(root=data_path, download=False, transform=transform)
elif dataset_name == 'cifar10':
    dataset = datasets.CIFAR10(root=data_path, train=True, download=True, transform=transform)
else:
    raise ValueError("Unsupported dataset.")

# Show dataset details
print(f"Dataset: {dataset_name}, total samples: {len(dataset)}")

# Show one image and label
img, label = dataset[0]
plt.imshow(img.permute(1, 2, 0))  # (C,H,W) → (H,W,C)
if dataset_name == 'sun397':
    plt.title(f"Label: {dataset.classes[label]}")
elif dataset_name == 'caltech101':
    plt.title(f"Label: {dataset.categories[label]}")
elif dataset_name == 'cifar10':
    plt.title(f"Label: {dataset.classes[label]}")
plt.axis("off")
plt.show()
```

# PCAM 

```python
import os
from torchvision import datasets, transforms
import matplotlib.pyplot as plt
import torch

# PCAM root directory
data_path = '/media/tom/datasets'

# Define transform
transform = transforms.Compose([
    transforms.Resize((96, 96)),  # optional, original is already 96x96
    transforms.ToTensor()
])

# Load PCAM dataset (train split)
dataset = datasets.PCAM(root=data_path, split='train', download=True, transform=transform)

# Show dataset stats
print(f"Total PCAM samples (train): {len(dataset)}")

# Visualize one sample
img, label = dataset[0]
plt.imshow(img.permute(1, 2, 0))  # CHW → HWC
plt.title(f"Label: {'Tumor' if label == 1 else 'Normal'}")
plt.axis("off")
plt.show()
```
